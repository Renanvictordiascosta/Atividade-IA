```python
import random
import numpy as np

pedidos = []
custos = []

i = 0

while i != 350:
  pedido = random.randint(1, 15)
  custo = random.randint(1, 20000)

  if i > 50:
    pedido = random.randint(2, 15)
    custo = random.randint(10000, 40000)

    if i >= 100:
      pedido = random.randint(5, 15)
      custo = random.randint(20000, 60000)

      if i >= 150:
        pedido = random.randint(10, 20)
        custo = random.randint(30000, 80000)

        if i >= 200:
          pedido = random.randint(15, 25)
          custo = random.randint(40000, 100000)

          if i >= 250:
            pedido = random.randint(20, 30)
            custo = random.randint(50000, 120000)

            if i >= 300:
              pedido = random.randint(25, 34)
              custo = random.randint(60000, 140000)

  pedidos.append(pedido)
  custos.append(custo)

  i=i+1

print(pedidos)
print(custos)
```

```python
i = 1

pedidos_x_custos= np.array([[pedidos[0], custos[0]]])

while i != 350:
  valor = np.array([pedidos[i], custos[i]])
  pedidos_x_custos = np.append(pedidos_x_custos, [valor], axis=0)
  i = i + 1

print(pedidos_x_custos)
```

```python
import matplotlib.pyplot as plot

plot.scatter(pedidos_x_custos[:,0], pedidos_x_custos[:,1], color="purple")

plot.xlim(0, 35)
plot.ylim(-5000, 145000)

plot.title("Gráfico Base", fontsize=15.0, fontweight='bold')

plot.xlabel("Número de Pedidos", fontsize=12.0, fontweight='bold')
plot.ylabel("Custo Total", fontsize=12.0, fontweight='bold')

plot.grid(True, color="black")

plot.show()
```

```python
from sklearn.cluster import KMeans

k_means = KMeans(
    n_clusters=3, init='random',
    n_init=10, max_iter=300,
    tol=1e-04, random_state=0
)

pred = k_means.fit_predict(pedidos_x_custos)

plot.scatter(pedidos_x_custos[:,0], pedidos_x_custos[:,1], c = pred)

plot.xlim(0, 35)
plot.ylim(-5000, 145000)

plot.scatter(k_means.cluster_centers_[:,0],k_means.cluster_centers_[:,1], s = 70, color = 'red')

plot.title("Resultado K-Means", fontsize=15.0, fontweight='bold')

plot.xlabel("Número de Pedidos", fontsize=12.0, fontweight='bold')
plot.ylabel("Custo Total", fontsize=12.0, fontweight='bold')

plot.grid(True, color="black")

plot.show()
```
